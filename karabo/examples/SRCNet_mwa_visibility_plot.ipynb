# MWA Visibility Waterfall Plots

This notebook demonstrates how to visualize MWA visibility data using waterfall plots.

It loads a UVFITS file and generates:
1. A time-frequency waterfall for a specific baseline (23×26, XX polarization)
2. FFT of all baselines binned by baseline length
3. FFT of a specific baseline (23×26, XX polarization)


## Download the UVFITS file

First, we need to download the MWA visibility data file.


```python
import os

uvfits_file = 'hyp_1184702048_ionosub_ssins_30l_src8k_300it_8s_80kHz_i1000.uvfits'

if not os.path.exists(uvfits_file):
    print(f"Downloading {uvfits_file}...")
    !wget https://projects.pawsey.org.au/high0.uvfits/hyp_1184702048_ionosub_ssins_30l_src8k_300it_8s_80kHz_i1000.uvfits
    print("Download complete!")
else:
    print(f"{uvfits_file} already exists.")
```

## Import required modules

Import the plotting function from the mwa_waterfall_baseline module.


```python
from mwa_waterfall_baseline import plot_waterfall

print("Imports successful!")
```

## Plot 1: Single baseline waterfall (23×26, XX polarization)

This produces a time-frequency waterfall plot for baseline 23×26 with XX polarization.


```python
from IPython.display import Image, display

plot_waterfall(
    uvfits_file,
    antpair=(23, 26),
    pol='XX',
    out='mwa_23x26_xx.png',
    show=True,
)

print("Plot saved as mwa_23x26_xx.png")
display(Image(filename='mwa_23x26_xx.png'))
```

## Plot 2: Single baseline FFT (23×26, XX polarization)

This produces an FFT waterfall plot for baseline 23×26 with XX polarization.


```python
plot_waterfall(
    uvfits_file,
    antpair=(23, 26),
    pol='XX',
    fft=True,
    out='mwa_23x26_fft_xx.png',
    show=True,
)

print("Plot saved as mwa_23x26_fft_xx.png")
display(Image(filename='mwa_23x26_fft_xx.png'))
```

## Plot 3: All baselines FFT (binned by baseline length)

This creates an FFT plot of all baselines, binned by baseline length (≤500m).


```python
plot_waterfall(
    uvfits_file,
    fft=True,
    all_baselines=True,
    max_baseline=500,
    bins=0,
    out='mwa_all.png',
    show=True,
)

print("Plot saved as mwa_all.png")

display(Image(filename='mwa_all.png'))
```
